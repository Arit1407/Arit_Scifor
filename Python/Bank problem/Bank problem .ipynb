```python
class BankAccount:
    def __init__(self,balance, pin):
        self.__balance = balance
        self.pin = pin

    def check_pin(self, input_pin):
        if self.pin == input_pin:
            return True
        else:
            print("Incorrect PIN.")
            return False

    def deposit(self, amount, input_pin):
        if self.check_pin(input_pin):
            self.__balance += amount
            print(f"Deposited {amount}. New balance: {self.__balance}")
        else:
            print("Deposit failed due to incorrect PIN.")

    def withdraw(self, amount, input_pin):
        if self.check_pin(input_pin):
            if amount <= self.__balance:
                self.__balance -= amount
                print(f"Withdrew {amount}. New balance: {self.__balance}")
            else:
                print("Insufficient balance.")
        else:
            print("Withdrawal failed due to incorrect PIN.")

    def get_balance(self, input_pin):
        if self.check_pin(input_pin):
            return self.__balance
        else:
            print("Failed to retrieve balance due to incorrect PIN.")
            return None

balance=int(input("Enter the Balance : "))
pin=input("Enter the atm pin : ")
account = BankAccount(balance,pin)
while True:
    print("\nChoose an option:")
    print("1.Deposit")
    print("2.Withdraw")
    print("3.Show balance")
    print("4. Exit")
    choice = input("Enter your choice: ")
    if choice == "1":
        deposit=int(input("Enter the deposite amount : "))
        deposit_pin=input("Enter the atm pin for deposite : ")
        account.deposit(deposit,deposit_pin)
    elif choice == "2":
        withdraw=int(input("Enter the withdraw amount : "))
        withdraw_pin=input("Enter the atm pin for withdraw : ")
        account.withdraw(withdraw,withdraw_pin)
    elif choice == "3":
        pin_for_balance=input("Enter the atm pin for balance : ")
        print(account.get_balance(pin_for_balance)) 
    elif choice == "4":
        print("Exiting...")
        break
    else:
        print("Invalid choice. Please try again.") 

```
