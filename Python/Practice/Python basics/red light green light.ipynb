```python
t=int(input())
for i in range(t):
    a,b=map(int,input().split())
    s=a+b
    sm=0
    l=list(map(int,str(s)))
    li=[6,2,5,5,4,5,6,3,7,6]
    for j in range(len(l)):
        for k in range(len(li)):
            if l[j]==k:
                sm+=li[k]
    print(sm)
```

```python
t=int(input())
for i in range(t):
    a,b=map(int,input().split())
    s=a+b
    sm=0
    l=list(map(int,str(s)))
    for j in range(len(l)):
        if l[j]==0 or l[j]==6 or l[j]==9:
            sm+=6
        elif l[j]==1:
            sm+=2
        elif l[j]==2 or l[j]==3 or l[j]==5:
            sm+=5
        elif l[j]==4:
            sm+=4
        elif l[j]==7:
            sm+=3
        else:
            sm+=7
    print(sm)
            
    
```
