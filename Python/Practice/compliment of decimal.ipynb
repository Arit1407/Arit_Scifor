```python
n=int(input("Enter the decimal number as a string:"))
nm=n
by=""
c_by=""
p=0
d=0
if n==0:
    print(1)
else:
    while n!=0:
        by+=str(n%2)
        n=n//2
    print(by[::-1])
    for a in by:
        if a=="0":
            c_by+="1"
        else:
            c_by+="0"
    print(c_by[::-1])
    for s in c_by:
        d+=int(s)*pow(2,p)
        p+=1
    print(d)
```
